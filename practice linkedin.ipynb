```python
def perm(df):
    return avocado_info["GrowthDuration"].sample(frac = 1).reset_index(drop = True)
```

```python
sim_test_stat = np.array([])
reps = 10000
for i in range(reps):
    perm_info = perm(avocado_info)
    df1 = pd.DataFrame({'Permuted Duration': perm_info, 'Fertilizer': avocado_info["Fertilizer"]})
    
    fertilized = df1.loc[df1['Fertilizer'] == True, 'Permuted Duration']
    not_fertilized = df1.loc[df1["Fertilizer"] == False, "Permuted Duration"]
    stat = np.mean(fertilized) - np.mean(not_fertilized)
    
    sim_test_stat = np.append(sim_test_stat, stat)
    
```

```python
fert = avocado_info.loc[avocado_info['Fertilizer'] == True,'GrowthDuration']
unfert = avocado_info.loc[avocado_info['Fertilizer'] == False,'GrowthDuration']
```

```python
fert,unfert
```

```python
np.random.choice(fert, size = len(fert))
```

```python
np.random.choice(unfert, size = len(unfert))
```

```python
def resample(data):
    return np.random.choice(data, size = len(data))
   
```

```python
def bootstrap(df2,reps):
    means = np.array([])
    for i in range(reps):
        new_sample = resample(df2)
        new_means = np.mean(new_sample)
        means = np.append(means, new_means)
        
    return means       
    
```

```python
fert_means = bootstrap (fert,10000)
unfert_means = bootstrap(unfert,10000)
estimate_means = fert_means - unfert_means
```

```python
estimate_means
```

```python
sns.distplot(estimate_means, kde = False)
plt.show
```
